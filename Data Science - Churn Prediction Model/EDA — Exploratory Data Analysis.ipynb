```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python
df = spark.table(
    "fiap_analytics.gold.rw_db_churn_ml_dataset"
)

display(df.limit(10))
```

```python
df.printSchema()
```

```python
churn_dist = df.groupBy("churn").count()

display(churn_dist)
```

## Distribuição do churn

Nesta análise verificamos a proporção entre:
- usuários que churnaram,
- usuários que permaneceram ativos.

Essa análise é importante para:
- entender o balanceamento do dataset,
- avaliar possíveis problemas de desbalanceamento,
- orientar a escolha das métricas do modelo.

```python
display(df.describe())
```

```python
sessions_analysis = df.groupBy("churn").agg(
    avg("total_sessions").alias("avg_sessions")
)

display(sessions_analysis)
```

```python
bet_analysis = df.groupBy("churn").agg(
    avg("avg_bet_amount").alias("avg_bet")
)

display(bet_analysis)
```

```python
recency_analysis = df.groupBy("churn").agg(
    avg("days_since_last_session").alias("avg_recency")
)

display(recency_analysis)
```

```python
pdf = df.toPandas()

import seaborn as sns
import matplotlib.pyplot as plt

corr = pdf.corr(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Matriz de Correlação")

plt.show()
```

```python
pdf["days_since_last_session"].hist(
    bins=30
)

plt.title(
    "Distribuição da Recência"
)

plt.show()
```
