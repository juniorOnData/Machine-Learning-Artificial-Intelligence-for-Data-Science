```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python
features_df = spark.table(
    "fiap_analytics.silver.rw_db_features_churn"
)

churn_df = spark.table(
    "fiap_analytics.silver.rw_db_dataset_churn"
)
```

```python
target_df = churn_df.groupBy("user_id").agg(
    max("churn").alias("churn")
)

final_ml_df = features_df.join(
    target_df,
    on="user_id",
    how="left"
)

final_ml_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "fiap_analytics.gold.rw_db_churn_ml_dataset"
    )

print("Dataset final de ML salvo com sucesso.")
```
